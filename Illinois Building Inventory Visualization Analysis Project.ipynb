# Relationship between the year acquired and the year constructed

The code that I use to create the following scatterplots are from Professor Naiman's Week 4 lecture code that can be accessed here https://github.com/UIUC-iSchool-DataViz/is445_spring2022/blob/master/week04/inClass_week04.ipynb. I changed the y-axis code in the original from "Square Footage" to "Year Constructed" in order to fit the assignment guidelines.

```python
buildings.plot(x="Year Acquired", y="Year Constructed", figsize=(20,6), rot=90, kind = 'scatter')
```

I attempted to create a scatterplot that depicts the relationship between the year acquired and year constructed, but clearly something is off. Why are some of the data points around year 0? That makes no sense. The state of Illinois and the United States in general didn't exist at the time, it was probably just Native American prairieland at the time. After examining the data, it turns out that there are many zeroes in the data for year acquired and year constructed. That is misleading, so I will turn these zeros into NaNs in a new dataframe. The following code for replacing zeros with NaNs is from Professor Naiman's Week 4 Prep Notebook that can be accessed here https://uiuc-ischool-dataviz.github.io/is445_spring2022/nbv.html?notebook_name=%2Fis445_spring2022%2Fweek04%2Fprep_notebook_week04.ipynb. I just changed the dataframe name from "b" in the original code to "buildings_no_zeros" because it is more specific and makes sense to me.

```python
buildings_no_zeros = pd.read_csv("https://uiuc-ischool-dataviz.github.io/is445_spring2022/week03/data/building_inventory.csv",
           na_values = {'Square Footage': 0,
                       'Year Acquired': 0,
                       'Year Constructed': 0}) 
buildings_no_zeros
```

Much better! Let's try to make the scatter plot again now that the misleading data has been dealt with.

```python
buildings_no_zeros.plot(x="Year Acquired", y="Year Constructed", figsize=(20,6), rot=90, kind = 'scatter')
plt.title('Relationship between the year acquired and the year constructed')
plt.show()
```

Now this is a proper data visualization for depicting the relationship between the year acquired and the year constructed.

# Median square footage in each Year Constructed as a function of Year Acquired

The following code's foundation is also from Professor Naiman's week 4 lecture code, similar to the groupby used in the second part of the assignment. I tried to take an approach similar to what Professor went over during lecture and also in the second part of the assignment, but it didn't really work out due to the presence of the third variable, the year constructed. Therefore, I have changed the code so that the median of all the columns are grouped by the year acquired, and I've also reset the index so that the values in this new dataframe can be used in the visualization afterwards.

```python
year_acquired_median = buildings_no_zeros.groupby("Year Acquired").median().reset_index()
year_acquired_median
```

The reason why I took the median of all the columns in this dataframe is because I needed the median of the square footage data and also some year data, but it doesn't matter if I take the median of a year, because the median of 2015 is still 2015 for example. The following code for the scatterplot and legend is compiled from multiple sources. I took the general skeleton of colored scatterplot code from this website https://stackoverflow.com/questions/59232073/scatter-plot-with-3-variables-in-matplotlib, and incorporated code from https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html in order to create the legend. 

```python
x = year_acquired_median["Year Acquired"]
y = year_acquired_median["Square Footage"]
z = year_acquired_median["Year Constructed"]

fig = plt.figure(figsize=(6, 6))
scatter = plt.scatter(x, y,
           linewidths=1, alpha=.7,
           edgecolor='k',
           s = 200,
           c=z)
plt.legend(*scatter.legend_elements(), loc="upper right", title="Year Constructed")
plt.title('Median square footage in each Year Constructed as a function of Year Acquired')
plt.xlabel('Year Acquired')
plt.ylabel('Median Square Footage')
plt.show()
```
