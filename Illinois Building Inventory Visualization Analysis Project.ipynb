```python
@ipywidgets.interact(feature1=ufosDS.columns, feature2=ufosDS.columns, colormap_name=plt.colormaps())
def create_scatter(feature1, feature2, colormap_name):
    with plt.style.context("ggplot"):
        fig = plt.figure(figsize=(12,6))
        
        plt.scatter(x = ufosDS[feature1],
                    y = ufosDS[feature2],
                    c = ufosDS[feature1],
                    cmap = colormap_name,
                    s = 30
                   )
        plt.xlabel(feature1.capitalize(), fontsize = 15)
        plt.ylabel(feature2.capitalize(), fontsize = 15)

        plt.title("%s vs %s"%(feature1.capitalize(), feature2.capitalize()))
        plt.xticks(rotation = 90, fontsize = 15)
        plt.yticks(fontsize = 15)
        plt.colorbar()
        plt.show()

#week 05 code from teacher
# @ipywidgets.interact(colormap_name=plt.colormaps(), take_log=True)
# def ufo_color(colormap_name, take_log=True):
#     if take_log:
#         norm = matplotlib.colors.SymLogNorm(10)
#     else:
#         norm = matplotlib.colors.Normalize()
#     plt.imshow(create_scatter(ufosDS.columns,ufosDS.columns), cmap=colormap_name, norm = norm)
#     plt.colorbar()
#     plt.clim(-352,352)
#     plt.show()
```
